# Advantages of Seaborn
1. Simplified Syntax
Seaborn reduces the amount of code needed to create complex visualizations. For example:
*   In Matplotlib, creating a basic scatter plot requires more lines of code.
*   In Seaborn, you can create the same plot with a single line.

2. Built-in Statistical Plots
Seaborn provides specialized functions for creating statistical plots that are not available in Matplotlib. These include:
*   Violin plots: Show distribution and density of data.
*   Box plots: Summarize distributions and identify outliers.
*   Pair plots: Visualize pairwise relationships in a dataset.
*   Heatmaps: Display correlation matrices or other matrix-like data.



3. Automatic Handling of Categorical Data
Seaborn makes it easy to work with categorical variables. For example:
*   It automatically groups data by categories and assigns colors.
*   It handles the placement of categorical data on axes.








```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the "tips" dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# same thing can be done using matplotlib
plt.scatter(tips["total_bill"], tips["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# plotiing lines
sns.lineplot(x="total_bill",y= "tip",data = tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip")
```

```python
# bar plot
sns.barplot(x="day",y = "total_bill",data = tips)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Scatter Plot of Total Bill in a Day")
```

```python
# histogram
sns.histplot(x="total_bill",data = tips,kde = True)
plt.title("Histogram of Total Bill")
#
```

```python
# box plot
sns.boxplot(x = "day", y = 'total_bill',data = tips)
plt.title("Box Plot of Total Bill Per day")
plt.show()
```

```python
# vilot plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
# Create a correlation matrix
# Select only numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr = numeric_tips.corr()
# print(corr)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
# Adding Hue for Categorical Data

sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Scatter Plot with Hue")
plt.show()
```

```python
# pair plot
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
# customized plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Customized Scatter Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```
