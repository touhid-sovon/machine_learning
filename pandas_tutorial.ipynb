The map() function in pandas is used to transform the values of a Series by applying a function, dictionary, or another Series. It is commonly used to replace values or modify them according to a specified rule.

When to Use map() vs apply()
map(): Ideal for element-wise transformations in a single column (Series).
apply(): Useful when you need more flexibility, like row- or column-wise transformations in a DataFrame.

```python
df1['first_name'].map({'John':'Jonathan','Jane':'Janice'})
```


Resampling in pandas refers to changing the frequency of time-series data by aggregating or interpolating. It’s commonly used to manipulate data with a DatetimeIndex and allows for operations like converting data from higher frequency (e.g., daily) to lower frequency (e.g., monthly) or vice versa.

Two types: Downsampling and Upsampling

```python
# resamoling
highs = df_time['High'].resample('D').max() # D for daily W for weekly
```
