```python
# plot number 1
plt.plot(ages_x,dev_y,'r', label  = 'All Devs')

# we can pass format strings format = '[marker][line][color]
#below the k is format string
# plot number 2
plt.plot(ages_x,py_dev_y,'k--' ,label = 'Python Devs')

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD)n By Age')
# plt.legend(['All Dev','Python Dev']) -- without explicitely giving legend names we can add the info when defining plt.plot
plt.legend()

plt.show()
```

Getting the percentage values and show them in the pieplot

```python
sum = 0
for num in slices:
  sum += int(num)

#one liner sum = sum(int(num) for num in slices)
print(sum)


percentages = []
for num in slices:
  percentages.append(round(int(num)/sum * 100, 2))

# oneliner
# percentages = [ round(int(num)/sum * 100,2) for num in slices]

print(percentages)
```

```python
# new_labels = []

# for text,num in list(zip(labels,percentages)):
#   val = text+'('+ str(num)+ '%)'
#   new_labels.append(val)

# one liner
new_labels = [f'{text}({num}%)' for text,num in zip(labels,percentages)]

print(new_labels[0])

```

```python
# plotting the new pie plot with percentages
plt.pie(slices,labels = new_labels,wedgeprops= {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Shortcut way of doing the same

```python
# Plot
plt.figure(figsize=(10, 8))
plt.pie(
    slices,
    labels=labels,
    autopct='%1.1f%%',  # Automatically calculates and displays percentages
    wedgeprops={'edgecolor': 'black'}  # Adds a border to slices
)

# Title and display
plt.title('Programming Languages by Usage')
plt.show()
```
